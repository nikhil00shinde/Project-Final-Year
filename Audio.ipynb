```python
from sklearn.metrics import precision_recall_curve
import matplotlib.pyplot as plt

# Example OCR output and ground truth
y_true = [0, 1, 0, 1] # binary labels for the ground truth (0 = negative, 1 = positive)
y_score = [0.8, 0.6, 0.7, 0.4] # scores or confidence levels for the OCR output

# Compute precision and recall for different thresholds
precision, recall, thresholds = precision_recall_curve(y_true, y_score)

# Plot the precision-recall curve
plt.plot(recall, precision, color='b', label='Tesseract OCR')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend(loc='lower left')

# Show the plot
plt.show()
```

```python
# Example data
TP = 85
FP = 5
FN = 10

# Calculate precision, recall, and F1 score
precision = TP / (TP + FP)
recall = TP / (TP + FN)
f1_score = 2 * (precision * recall) / (precision + recall)

# Print the results
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1_score)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import precision_recall_curve

# Define the data
TP = 85
FP = 5
FN = 10

# Calculate precision and recall
precision = TP / (TP + FP)
recall = TP / (TP + FN)

# Create arrays with a single value for precision and recall
precision_array = np.array([precision])
recall_array = np.array([recall])

# Create a plot with a single point
plt.plot(recall_array, precision_array, marker='o', markersize=10)
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')

# Add text annotations for the point
plt.annotate(f'TP = {TP}\nFP = {FP}\nFN = {FN}', xy=(recall, precision),
             xytext=(recall + 0.1, precision + 0.1),
             arrowprops=dict(facecolor='black', shrink=0.05))

# Set the axis limits
plt.xlim([0, 1])
plt.ylim([0, 1])

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

# Define the data
TP = 85
FP = 5
FN = 10

# Create the confusion matrix
conf_matrix = np.array([[TP, FP], [FN, 0]])

# Plot the confusion matrix
fig, ax = plt.subplots()
im = ax.imshow(conf_matrix)

# Add axis labels and tick marks
ax.set_xticks([0, 1])
ax.set_yticks([0, 1])
ax.set_xticklabels(['Predicted Positive', 'Predicted Negative'])
ax.set_yticklabels(['Actual Positive', 'Actual Negative'])
ax.tick_params(axis='both', labelsize=10)

# Add colorbar
cbar = ax.figure.colorbar(im, ax=ax)

# Loop over data dimensions and create text annotations
for i in range(2):
    for j in range(2):
        text = ax.text(j, i, conf_matrix[i, j], ha='center', va='center', color='w', fontsize=20)

# Add title
ax.set_title('Confusion Matrix', fontsize=16)

# Show the plot
plt.show()
```
